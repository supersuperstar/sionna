```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#read and turn into float list
path = 'model/0409/'
train_loss = np.loadtxt(path+"train_loss.txt")
train_acc = np.loadtxt(path+"train_acc.txt")
loss = np.loadtxt(path+"loss.txt")
acc = np.loadtxt(path+"acc.txt")
```

```python
plt.plot(np.arange(len(loss)), loss, label='loss')
plt.plot(np.arange(len(train_loss)), train_loss, label='train_loss')
plt.plot(np.arange(len(acc)), acc, label='acc')
plt.plot(np.arange(len(train_acc)), train_acc, label='train_acc')
```

```python
epochs = range(1, len(loss) + 1) 
# 创建图和轴
fig, ax1 = plt.subplots()

# 设置x轴为对数刻度
ax1.set_xscale('log')

# 绘制loss曲线
color = 'tab:red'
ax1.set_xlabel('Epoch')
ax1.set_ylabel('Loss', color=color)
ax1.plot(epochs, loss, 'r--', label='Validation Loss')  # 红色虚线
ax1.plot(epochs, train_loss, 'r-', label='Training Loss')  # 红色实线
ax1.tick_params(axis='y', labelcolor=color)

# 创建另一个y轴用于acc曲线
ax2 = ax1.twinx()  
color = 'tab:blue'
ax2.set_ylabel('Accuracy', color=color)
ax2.plot(epochs, acc, 'b-', label='Validation Accuracy')  # 蓝色实线
ax2.plot(epochs, train_acc, 'b--', label='Training Accuracy')  # 蓝色虚线
ax2.tick_params(axis='y', labelcolor=color)

# 添加图例
lines, labels = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax2.legend(lines + lines2, labels + labels2, loc='center right', fancybox=True, shadow=True, ncol=1)

# 显示图表
plt.show()
```
